```python
from sklearn import tree
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import graphviz
```

```python
df = pd.read_csv("Couples.csv")
label_encoder = LabelEncoder()
df['GroomMB'] = label_encoder.fit_transform(df['GroomMB'])
df['BrideMB'] = label_encoder.fit_transform(df['BrideMB'])
```

```python
x_train = df[["GroomMB","BrideMB","GroomInc","BrideInc"]]
Y_train = df["Outcome"]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, Y_train)
```

```python
tree.plot_tree(clf)
```

```python
data1 = tree.export_graphviz(clf, out_file= None)
graph = graphviz.Source(data1)
graph.render("decision_tree")
```
